# <span style="color:red">Introduction and Motivation </span>


### What is "data science"? 

A discipline that concerns itself with

*  Data collection
*  Data storage, retrieval, and curation
*  Data description and visualization
*  Statistical inference
*  Machine learning
*  Design of computational tools for these activities

Data science draws on other disciplines, in particular computer science and statistics.


### What are the goals of "Introduction to Data Science"?
 
1.  Teach the fundamentals of data science
2.  Provide an opportunity to gather hands-on experience 

to students who had no previous exposure to programming or statistics.

**Data are ubiquitous, and therefore analyzing data is an important skill.** 

Businesses collect data
*  Barcode scanners in stores
*  Click data from Web sites
*  Product ratings (mention recommender systems)
*  Loan compliance
*  etc

[Governments on all levels collect data](https://www.data.gov/). Rational goverment is impossible without data.

Researchers in all fields collect data
*  Astronomy: [Sloan digital sky survey](http://www.sdss.org/), [Large Synoptic Survey Telescope](https://www.lsst.org/) (15TB per night)
*  Physics: [Large Hadron collider](https://home.cern/topics/large-hadron-collider)
*  Biology: [Ecological field work](http://ecosystemsentinels.org/)
*  Medicine: [Human Genome Project](https://www.genome.gov/10001772/all-about-the--human-genome-project-hgp/), proteomics

Data collection and analysis is at the very heart of science.
