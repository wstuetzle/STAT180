```python
baby = Table.read_table("baby.csv")
baby.show(10)
```

```python
baby.scatter("Maternal Pregnancy Weight", "Birth Weight", width = 8, height = 6, s = 0.2)
```

```python
baby.scatter("Maternal Age", "Birth Weight", width = 8, height = 6, s = 0.2)
```

```python
baby.hist("Gestational Days", bins = np.arange(200, 370, 10))
```

```python
baby_sub = baby.where("Gestational Days", are.between(230, 320))

```

```python
baby_sub.scatter("Gestational Days", "Birth Weight", width = 8, height = 6, s = 0.4)
```

```python
baby.scatter("Gestational Days", "Birth Weight", width = 8, height = 6, s = 0.2)
```

```python
baby.scatter("Maternal Height", "Maternal Pregnancy Weight", width = 8, height = 6, s = 0.2)
```

```python
baby_smoker = baby.where("Maternal Smoker", are.equal_to(True))
baby_nonsmoker = baby.where("Maternal Smoker", are.equal_to(False))
baby_smoker.hist("Birth Weight", bins = np.arange(50, 190, 10))
baby_nonsmoker.hist("Birth Weight", bins = np.arange(50, 190, 10))
print(np.mean(baby_smoker.column("Birth Weight")))
print(np.mean(baby_nonsmoker.column("Birth Weight")))
```
